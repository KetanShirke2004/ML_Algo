```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn as sns
```

```python
df=pd.read_csv("placement.csv")
df
```

```python
X=df.iloc[:,0]
y=df.iloc[:,-1]
```

```python
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
# scikit-learn expects a 2D array for X; convert Series to a single-column array/DataFrame
reg.fit(X.values.reshape(-1, 1), y)
```

```python
m=reg.coef_
```

```python
b=reg.intercept_
```

```python
plt.scatter(X, y)
# ensure X passed to predict is 2D
X_2d = X.values.reshape(-1, 1)
plt.plot(X, reg.predict(X_2d), color='red')
```

```python
#lets apply gradient descent assuming slop os constant m=0.56959129
#and lets assume the starting value for intercept b=0
y_pred=((0.56959129*X)+0)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X_2d),color='red',label='OLS')
plt.plot(X,y_pred,color='blue',label='GD')
plt.legend()
plt.show()
```

```python
m=reg.coef_
b=reg.intercept_
```

```python
loss_slope=-2*np.sum(y-m*X.ravel()-b)
loss_slope
```
