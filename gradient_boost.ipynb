```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X=np.random.rand(100,1)-0.5
y=3*X[:,0]**2+0.05*np.random.randn(100)
```

```python
import pandas as pd 
```

```python
df=pd.DataFrame()
```

```python
df['X']=X.reshape(100)
df['y']=y
```

```python
plt.scatter(df['X'],df['y'])
plt.title('Scatter plot of X vs y')
```
